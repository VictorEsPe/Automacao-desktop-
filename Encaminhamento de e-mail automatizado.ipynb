Automação de Sistemas e Processos com Python
Desafio:
Todos os dias, o nosso sistema atualiza as vendas do dia anterior. O seu trabalho diário, como analista, é enviar um e-mail para a diretoria, assim que começar a trabalhar, com o faturamento e a quantidade de produtos vendidos no dia anterior

E-mail da diretoria: seugmail+diretoria@gmail.com
Local onde o sistema disponibiliza as vendas do dia anterior: https://drive.google.com/drive/folders/149xknr9JvrlEnhNWO49zPcw0PW5icxga?usp=sharing

Para resolver isso, vamos usar o pyautogui, uma biblioteca de automação de comandos do mouse e do teclado

```python
import pyautogui
import pyperclip
import time

pyautogui.PAUSE = 1

#entrar no sistema usando o link https://drive.google.com/drive/folders/149xknr9JvrlEnhNWO49zPcw0PW5icxga
pyautogui.hotkey("ctrl","t")
pyperclip.copy("https://drive.google.com/drive/folders/149xknr9JvrlEnhNWO49zPcw0PW5icxga")
pyautogui.hotkey("ctrl","v")
pyautogui.press("enter")
time.sleep(5)

#entrar no aba "exportar" e baixar o arquivo
pyautogui.click(x=453, y=278, clicks=2) #abre a pasta "exportar"

pyautogui.click(x=432, y=323) #seleciona o arquivo
pyautogui.click(x=1167, y=163) #clica nos 3 pontinhos
time.sleep(2)
pyautogui.click(x=973, y=572) #baixar o arquivo 
time.sleep(5)
```

```python
import pandas as pd

# importar a tabela
tabela = pd.read_excel(r"C:\Users\Usuario\Downloads\Vendas - Dez.xlsx") #Lê o arquivo \ #Deve mostrar o caminho para o arquivo. Ex.:C:\Users\Usuario\Downloads \ #A letra "r" mostra ao programa que ele deve ler o arquivo mencionado
display(tabela) #esse comando exibe as informações

# calcular o faturamento total e a quantidade de produtos vendidos
faturamento = tabela["Valor Final"].sum() # seleciona a coluna citada e soma seus valores
qtde_produtos = tabela["Quantidade"].sum()
display(faturamento)
display(qtde_produtos)
```

```python
# abrir email
pyautogui.hotkey("ctrl","t")
pyperclip.copy('https://mail.google.com/mail/u/0/#inbox')
pyautogui.hotkey('ctrl','v')
pyautogui.press('enter')
time.sleep(5)

#escrever email
pyautogui.click(x=85, y=172) # Clica em "escrever"
time.sleep(3)
pyautogui.write('s')
pyautogui.press("tab") # seleciona o email correspondente
pyautogui.press("tab") # muda para o campo assunto
pyperclip.copy('Relatório de Vendas') 
pyautogui.hotkey('ctrl','v')
pyautogui.press('tab') # muda para o campo do texto

# Para escrever textos que possuam mais de uma linha basta inserir três aspas simples ou duplas no começo e no final do texto
# A letra "f" diz ao programa que queremos formatar o texto com variáveis
texto = f'''
Prezados, bom dia

Faturamento foi de R${faturamento:,.2f}
A quantidade de produtos vendidos foi de {qtde_produtos:,}

Victor
'''

#enviar 
pyperclip.copy(texto)
pyautogui.hotkey('ctrl','v')
pyautogui.hotkey('ctrl','enter')
```
